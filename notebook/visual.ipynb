Visualisations

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd 
```

```python
df= pd.read_csv("loan_portfolio.csv")
```

```python
df.head()
```

FICO Score distribution

```python
fico_filtered = df[(df["FICO"] > 580) & (df["FICO"] < 820)]
```

```python

sns.set_style("whitegrid")

plt.figure(figsize=(8,5))

sns.histplot(fico_filtered["FICO"], bins=30, kde=True, color="steelblue")

plt.title("FICO Score Distribution")
plt.xlabel("FICO Score")
plt.ylabel("Number of Borrowers")

plt.show()
```

LTV Histogram

```python
plt.figure(figsize=(8,5))

sns.histplot(df["LTV"], bins=30, kde=True, color="darkorange")
plt.axvline(df["LTV"].mean(), linestyle="--", label="Average LTV")

plt.title("Loan-to-Value (LTV) Distribution")
plt.xlabel("Loan-to-Value Ratio")
plt.ylabel("Number of Loans")

plt.show()
```

Loan Amount Distribution

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8,5))

sns.histplot(df["Loan_Amount"], bins=30, kde=True, color="steelblue")

plt.title("Distribution of Mortgage Loan Amounts")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")

plt.show()
```

PD vs LTV scatter

```python

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="LTV",
    y="PD_base",
    alpha=0.6,
    hue="FICO",
    palette="viridis",
    
)

plt.title("Probability of Default vs Loan-to-Value Ratio")
plt.xlabel("Loan-to-Value (LTV)")
plt.ylabel("Probability of Default (PD)")

plt.show()
```
